```python
glulamrawdata= pd.read_csv('glulam_data5.csv')
```

#LinearRegression Forecasting

```python
glulamMargin_traina_timeseries1= TimeSeries.from_series(glulamrawdata['Margin_Stattransformed'])
glulamMargin_traina_timeseries= glulamMargin_traina_timeseries1[:-30]
```

```python

model=RegressionModel(
                       lags=[-1, -4, -12],
                       model=LinearRegression()
                       )

model.fit(series=glulamMargin_traina_timeseries)


glulamMargin_traina_timeseries_pred=model.predict(n=30, series=glulamMargin_traina_timeseries)
glulamMargin_traina_timeseries1.plot()

glulamMargin_traina_timeseries_pred.plot(label='forecast', low_quantile=0.05, high_quantile=0.95)
plt.legend()
```
